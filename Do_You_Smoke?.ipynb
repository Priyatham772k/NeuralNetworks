##Remove Outliers

```python
import pandas as pd
import numpy as np


data_no_outliers_df = pd.DataFrame()

# List of columns to exclude from outlier removal
exclude_columns = ['sex', 'hear_left', 'hear_right', 'urine_protein']  # Replace with your actual column names

for column in final_data.columns:
    if column in exclude_columns:

        data_no_outliers_df[column] = final_data[column]
    else:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = final_data[column].quantile(0.25)
        Q3 = final_data[column].quantile(0.75)

        # Calculate the IQR (Interquartile Range)
        IQR = Q3 - Q1

        # Determine the outlier range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR


        filtered_column = final_data[column][(final_data[column] >= lower_bound) & (final_data[column] <= upper_bound)]


        data_no_outliers_df = pd.concat([data_no_outliers_df, filtered_column], axis=1).dropna()


```





2. Chi-Square Test for Categorical Features

The Chi-Square test is used to determine if there is a significant association between two categorical variables. It's suitable for your categorical features like 'sex', 'hear_left', 'hear_right', and 'urine_protein'.





```python
from scipy.stats import chi2_contingency
import pandas as pd

categorical_features = ['sex', 'hear_left', 'hear_right', 'urine_protein']
chi2_results = pd.DataFrame(columns=['Feature', 'Chi2 Value', 'p-value'])

for feature in categorical_features:
    contingency_table = pd.crosstab(final_data[feature], final_data['SMK_stat_type_cd'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    chi2_results = chi2_results.append({'Feature': feature, 'Chi2 Value': chi2, 'p-value': p}, ignore_index=True)

print(chi2_results)
```

3.ANOVA for Numerical Features

ANOVA is used to compare the means of a numerical feature across different groups defined by a categorical target variable.

```python
from scipy.stats import f_oneway

numerical_features = ["age", "height", "weight", "waistline", "sight_left", "sight_right",
                      "SBP", "DBP", "BLDS", "tot_chole", "HDL_chole",
                      "LDL_chole", "triglyceride", "hemoglobin", "serum_creatinine",
                      "SGOT_AST", "SGOT_ALT", "gamma_GTP"]
anova_results = pd.DataFrame(columns=['Feature', 'F Value', 'p-value'])

for feature in numerical_features:
    groups = final_data.groupby('SMK_stat_type_cd')[feature].apply(list)
    f_val, p_val = f_oneway(*groups)
    anova_results = anova_results.append({'Feature': feature, 'F Value': f_val, 'p-value': p_val}, ignore_index=True)

print(anova_results)
```

1.Mean and Median Imputation Method


```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt


categorical_cols = ['sex', 'hear_left', 'hear_right', 'urine_protein', 'SMK_stat_type_cd', 'DRK_YN']
numeric_cols = ["age", "height", "weight", "waistline", "sight_left", "sight_right",
                "SBP", "DBP", "BLDS", "tot_chole", "HDL_chole",
                "LDL_chole", "triglyceride", "hemoglobin", "serum_creatinine",
                "SGOT_AST", "SGOT_ALT", "gamma_GTP"]

# Function to remove a percentage of data
def remove_data(df, percent, categorical_cols, numeric_cols):
    modified_df = df.copy()
    for col in modified_df.columns:
        if col in categorical_cols:
            mode = modified_df[col].mode()[0]
            modified_df.loc[modified_df.sample(frac=percent).index, col] = mode
        elif col in numeric_cols:
            modified_df.loc[modified_df.sample(frac=percent).index, col] = np.nan
    return modified_df

modified_data_1 = remove_data(final_data, 0.01, categorical_cols, numeric_cols)
modified_data_5 = remove_data(final_data, 0.05, categorical_cols, numeric_cols)
modified_data_10 = remove_data(final_data, 0.10, categorical_cols, numeric_cols)

# Imputers
mean_imputer = SimpleImputer(strategy='mean')
median_imputer = SimpleImputer(strategy='median')

# Function to apply imputation and calculate errors
def impute_and_evaluate(data, imputer, original, numeric_cols):
    imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
    error = calculate_percentage_error(original, imputed_data, numeric_cols)
    return imputed_data, error

# Function to calculate percentage error for numeric columns
def calculate_percentage_error(original, imputed, numeric_cols):
    original_numeric = original[numeric_cols]
    imputed_numeric = imputed[numeric_cols]
    error = np.nanmean(np.abs((original_numeric - imputed_numeric) / original_numeric) * 100)
    return np.nan_to_num(error)

# Apply imputation and calculate errors for mean imputation
imputed_data_1_mean, error_1_mean = impute_and_evaluate(modified_data_1, mean_imputer, final_data, numeric_cols)
imputed_data_5_mean, error_5_mean = impute_and_evaluate(modified_data_5, mean_imputer, final_data, numeric_cols)
imputed_data_10_mean, error_10_mean = impute_and_evaluate(modified_data_10, mean_imputer, final_data, numeric_cols)


print("Mean Imputation Errors:")
print(f"1% Data Removed Error: {error_1_mean}")
print(f"5% Data Removed Error: {error_5_mean}")
print(f"10% Data Removed Error: {error_10_mean}")


imputed_data_1_median, error_1_median = impute_and_evaluate(modified_data_1, median_imputer, final_data, numeric_cols)
imputed_data_5_median, error_5_median = impute_and_evaluate(modified_data_5, median_imputer, final_data, numeric_cols)
imputed_data_10_median, error_10_median = impute_and_evaluate(modified_data_10, median_imputer, final_data, numeric_cols)


print("\nMedian Imputation Errors:")
print(f"1% Data Removed Error: {error_1_median}")
print(f"5% Data Removed Error: {error_5_median}")
print(f"10% Data Removed Error: {error_10_median}")

def calculate_bias_variance(original, imputed, numeric_cols):
    bias_variance_df = pd.DataFrame(columns=['Feature', 'Bias', 'Variance'])

    for col in numeric_cols:
        original_col = original[col].dropna()
        imputed_col = imputed[col]
        bias = np.mean(imputed_col - original_col)
        variance = np.var(imputed_col - original_col)

        bias_variance_df = bias_variance_df.append({
            'Feature': col,
            'Bias': bias,
            'Variance': variance
        }, ignore_index=True)

    return bias_variance_df

# Calculate bias and variance for mean imputation
bias_variance_mean_1 = calculate_bias_variance(final_data, imputed_data_1_mean, numeric_cols)
bias_variance_mean_5 = calculate_bias_variance(final_data, imputed_data_5_mean, numeric_cols)
bias_variance_mean_10 = calculate_bias_variance(final_data, imputed_data_10_mean, numeric_cols)

# Calculate bias and variance for median imputation
bias_variance_median_1 = calculate_bias_variance(final_data, imputed_data_1_median, numeric_cols)
bias_variance_median_5 = calculate_bias_variance(final_data, imputed_data_5_median, numeric_cols)
bias_variance_median_10 = calculate_bias_variance(final_data, imputed_data_10_median, numeric_cols)


print("Bias and Variance for Mean Imputation (1% Data Removed):")
print(bias_variance_mean_1)
print("\nBias and Variance for Mean Imputation (5% Data Removed):")
print(bias_variance_mean_5)
print("\nBias and Variance for Mean Imputation (10% Data Removed):")
print(bias_variance_mean_10)

print("\nBias and Variance for Median Imputation (1% Data Removed):")
print(bias_variance_median_1)
print("\nBias and Variance for Median Imputation (5% Data Removed):")
print(bias_variance_median_5)
print("\nBias and Variance for Median Imputation (10% Data Removed):")
print(bias_variance_median_10)

```

2.Mode Imputation Method


```python
from sklearn.impute import SimpleImputer

# Mode imputer
mode_imputer = SimpleImputer(strategy='most_frequent')

# Function to apply imputation and calculate errors
def impute_and_evaluate(data, imputer, original, numeric_cols):
    imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
    error = calculate_percentage_error(original, imputed_data, numeric_cols)
    return imputed_data, error

# Apply mode imputation and calculate errors
imputed_data_1_mode, error_1_mode = impute_and_evaluate(modified_data_1, mode_imputer, final_data, numeric_cols)
imputed_data_5_mode, error_5_mode = impute_and_evaluate(modified_data_5, mode_imputer, final_data, numeric_cols)
imputed_data_10_mode, error_10_mode = impute_and_evaluate(modified_data_10, mode_imputer, final_data, numeric_cols)

# Print the mode imputation errors
print("Mode Imputation Errors:")
print(f"1% Data Removed Error: {error_1_mode}")
print(f"5% Data Removed Error: {error_5_mode}")
print(f"10% Data Removed Error: {error_10_mode}")

# Calculate bias and variance for mode imputation
bias_variance_mode_1 = calculate_bias_variance(final_data, imputed_data_1_mode, numeric_cols)
bias_variance_mode_5 = calculate_bias_variance(final_data, imputed_data_5_mode, numeric_cols)
bias_variance_mode_10 = calculate_bias_variance(final_data, imputed_data_10_mode, numeric_cols)

# Print out the bias and variance for mode imputation
print("Bias and Variance for Mode Imputation (1% Data Removed):")
print(bias_variance_mode_1)
print("\nBias and Variance for Mode Imputation (5% Data Removed):")
print(bias_variance_mode_5)
print("\nBias and Variance for Mode Imputation (10% Data Removed):")
print(bias_variance_mode_10)
```
